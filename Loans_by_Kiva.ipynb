# Visualizing Loans Awarded by Kiva


## Step 1: Import Necessary Python Modules


```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

## Step 2: Ingest The Data


```python
df=pd.read_csv('kiva_data.csv')
```

## Step 3: Examine The Data


```python
print(df.head(25))
```

## Step 4: Bar Charts

```python
sns.set_style('whitegrid')
f, ax = plt.subplots(figsize=(15, 10))

sns.barplot(data = df, x= "country", y= "loan_amount")
```

```python
import matplotlib.ticker as mtick
f, ax = plt.subplots(figsize=(15, 10))
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick) 

sns.barplot(data=df, x="country", y = "loan_amount")
plt.show()
```

## Step 5: Learn More By Using `hue`

```python
f, ax = plt.subplots(figsize=(15, 10))


fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

sns.barplot(data=df, x="country", y = "loan_amount", hue="gender")
#plt.show()
```

```python
# Observations

# On average, male recepients receive larger loans from Kiva.

# Across the five countries, El Salvador appears to have the least disparity in loan amounts (by gender).

# Based on the data, Kiva should work to reduce gender loan gap. 

### Recommendations

# Kiva could hold workshops focused on women-led projects to reduce loan gap
# Kiva could require that an equal amount of loans be given to male and female-driven projects
```

## Step 6: Styling





```python
sns.set_palette("Dark2")

sns.set_style("whitegrid")

plt.figure(figsize=(20, 15))

ax.set_title("Loans given by Kiva for each country")
sns.barplot(data=df, x="country", y="loan_amount", hue="gender")
plt.show()
```

## Step 7: Box Plots With Kiva Data



```python
plt.figure(figsize=(16, 10))
sns.boxplot(data=df, x="country", y="loan_amount")
plt.show()
```

```python
# Kenya has the widest range of loan amounts of all the countries
# The average highest loan amounts were aawarded in Cambodia
```

```python
plt.figure(figsize=(16, 10))

sns.boxplot(data=df, x="activity", y="loan_amount")
plt.show()
```

```python
# On average, "Farming" receives the largest loan amount and also has the widest distribution.
```

## Step 9: Violin Plots

```python
plt.figure(figsize=(16, 10))

sns.violinplot(data=df, x="activity", y = "loan_amount")
plt.show()
```

```python
plt.figure(figsize=(16, 10))

sns.violinplot(data=df, x="country", y = "loan_amount")
plt.show()
```

## Step 10: Split Violin Plots

```python
sns.set_palette("Spectral")
plt.figure(figsize=(18, 12))

sns.violinplot(data=df, x="country", y = "loan_amount", hue="gender", split=True)
plt.show()
```

```python
# In Pakistan, the average loan amount received by women is much less than the average loan amount received by men.
# The same can be said for Cambodia (more or less).
# In El Salvador, there is a fairly equal distribution of loans (total number of loans) between genders, as well as
# similar median loan amounts (on average).
```
