```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('all_data.csv')
print(df.head())
```

```python
print(df["Country"].nunique())
```

```python
# There are six countries represented in the data
```

```python
print(df["Year"].nunique())
```

```python
# There are sixteen different years represented in the data
```

```python
df.rename(columns = {'Life expectancy at birth (years)': 'LEABY'}, inplace=True)

print(df.head())
```

```python
f, ax = plt.subplots(figsize=(15, 10))
plt.bar(df['Country'], df['GDP'])

plt.show()
```

***


```python
f, ax = plt.subplots(figsize=(15, 10))

plt.bar(df['Country'], df['LEABY'])

plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(14, 12))

sns.violinplot(data=df, x="Country", y = "LEABY")
plt.show()
```

```python
#The life expectancy at birth has changed the most in Zimbabwe
```

```python
sns.set_palette("Dark2")

sns.set_style("whitegrid")

plt.figure(figsize=(20, 15))

ax = sns.barplot()
ax = sns.barplot(x="Country", y="GDP", hue="Year", data=df)


plt.ylabel("GDP in Trillions of U.S. Dollars")

plt.show()
```

```python
sns.set_palette("Dark2")

sns.set_style("whitegrid")

plt.figure(figsize=(20, 15))

ax = sns.barplot()
ax = sns.barplot(x="Country", y="LEABY", hue="Year", data=df)


plt.ylabel("LEABY")

plt.show()
```

```python
plt.figure(figsize=(20, 15))
g = sns.FacetGrid(df, col= "Year", hue="Country", col_wrap=4, height=2)
g = (g.map(plt.scatter, "GDP", "LEABY", edgecolor="w").add_legend())

plt.show()
```

```python
# The United States of America had the highest GDPs in all the years recorded in the data.
# Germany had the highest life expectancy at birth on average across all the years recorded in the data.
```

```python
g3 = sns.FacetGrid(df, col="Country", col_wrap=3, height=4)
g3 = (g3.map(plt.plot, "Year", "LEABY").add_legend())

plt.show()
```
